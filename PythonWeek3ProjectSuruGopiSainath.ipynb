# 3Q. Check for the Percentage of the missing values and drop or impute them.


```python
Missing_values_percentage = round(((data_df.isna().sum()/len(data_df))*100),2)
print(Missing_values_percentage)

# Here, data_df.isna().sum() gives count of all missing values in there respective columns.
# (data_df.isna().sum()/len(data_df))*100 gives percentage of missing values to the total count of column.
```

```python
# By the above output, we can say that there is no missing values in 11 columns.
# Only 1 column that is "comments" column has 0.42 % missing values in it.
# As it's a very less percentage of missing values, we can impute them.
# There is no importance of comments data in next questions but if we drop all rows having missing values in comments,
# leads to loss of useful information of other columns which may useful in next questions.
# So, we impute missing values of 'comments' as " " (empty) using .fillna() method.

data_df1 = data_df.fillna(" ")
data_df1.isna().sum()

# we can see below output that there is no more missing values in any column.
```

# 5Q. Convert datatype of column published_at from object to pandas datetime.

```python
data_df1['published_at'] = pd.to_datetime(data_df1['published_at'])

# we use pd.to_datetime() method to convert 'published_at' column datatype from 'object' to pandas 'datetime' datatype.
```

```python
data_df1.dtypes

# from below output we can say that 'published_at' column converted to 'datetime' datatype.
```

# 6Q. Create a new column as 'published_month' using the column published_at (display the months only)

```python
data_df1['published_month'] = data_df1['published_at'].dt.month
print(data_df1['published_month'])

# here 'published_at' column is datetime datatype, so we have a method .dt.month in it to extract month.
# from below output we can say that new column 'published_month' is created and had the values of months of 'published_at' column.
```

# 7Q. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....

```python
month_dict={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
data_df1['published_month'].replace(month_dict,inplace=True)
print(data_df1['published_month'])

# we used 'month_dict' dictionary to declare month names with respect to there number.
# by using .replace method we replaced 'published_month' column from number to there respective month name.
# in below output we can see that numbers replaced with month names to whole 'published_month' column.
```

# 8Q. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
data_df1.groupby('published_month')['video_id'].count().to_frame().sort_values(by='video_id',ascending=False)

# we grouped 'published_month' with count of 'video_id' and sorted in decreasing order of count of 'video_id'
```

# 9Q. Find the count of unique video_id, channel_id and channel_title.

```python
print("count of unique video_id : ",data_df1['video_id'].nunique())
print("count of unique channel_id : ",data_df1['channel_id'].nunique())
print("count of unique channel_title : ",data_df1['channel_title'].nunique())

# used .nunique() method to extract count of unique values
```

# 10Q. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.


```python
print("\n top 10 channel names having highest number of videos are: ")
data_df1.groupby('channel_title')['video_id'].count().to_frame().sort_values(by='video_id',ascending=False).head(10)

# used .groupby method to group 'channel_title' with count of 'video_id' and sorted in decreasing order of count of 'video_id'
# and extracted top 10 records using .head method
```

```python
print("\n bottom 10 channel names having lowest number of videos are: ")
data_df1.groupby('channel_title')['video_id'].count().to_frame().sort_values(by='video_id',ascending=False).tail(10)

# used .groupby method to group 'channel_title' with count of 'video_id' and sorted in decreasing order of count of 'video_id'
# and extracted bottom 10 records using .tail method
```

# 11Q. Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.

# 13Q. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


```python
Correlation=data_df1['view_count'].corr(data_df1['dislikes'])
print("\n correlation between view_count and dislikes is:  ",round(Correlation,2),"\n")

# used .corr method to see correlation between 'view_count','dislikes'.
```

```python
plt.scatter(data_df1['view_count'],data_df1['dislikes'],c='b')
plt.title("\n Scatter plot of view_count,dislikes \n")
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.show()

# used matplotlib.pyplot to plot a scatter plot.
# scatter plot explains correlation between 2 variables.
```

## **<font color="blue">Inferences:</font>**
* Correlation is mutual relationship between 2 columns.
* Here, correlation between view_count, dislikes is 0.68. 
* As it is 0.68, we can conclude that there is positive correlation between view_count,dislikes.
* that means, if view_count value increases, dislikes value also increases.
* if view_count value decreases, dislikes value also decreases.
* Here in scatter plot, we can easily see that as view_count values increases, dislikes values also increasing.
